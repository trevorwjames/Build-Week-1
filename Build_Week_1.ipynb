```
# parsing out an array of neighbourhoods to create a 
# new df for average price per AirBnB in manhattan neighbourhoods
hoods = manhattan['neighbourhood'].unique()
```

```
hoods
```

```
# for loop assigning price averages per neighbourhood
avgprice_hood = {}

for hood in hoods:
  avgprice_hood[hood] = [manhattan[manhattan['neighbourhood'] == hood]['price'].mean()]
print(avgprice_hood)
```

```
# data frame of avg price per neighbourhood in manhattan
# rounded the price to 1 decimal to make graphing more readable

dfm = pd.DataFrame(avgprice_hood).T.reset_index().round(1)

dfm.columns=['neighborhood', 'avg_price']

print(dfm.shape)
dfm.head()
```

```
# using altair to build visualizations
import altair as alt
```

```
# first looking at all of manhattan to find the most reasonable spot to stay

bars = alt.Chart(dfm, title='Average AirBnB Price per night for different Neighborhoods in Manhattan').mark_bar().encode(
    y=alt.Y('neighborhood', axis=alt.Axis(title='Manhattan Neighborhood')),
    x=alt.X('avg_price', axis=alt.Axis( title='Average Price ($)')),
    color=alt.condition(alt.datum.neighborhood == 'Tribeca', 
                        alt.value('red'),
                        alt.value('blue')
                        )
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text='avg_price'
)

bigchart = (bars + text)

bigchart
```

```
# was thinking of also creating a visualization for average price per borough
# would be too repetiive for size of post, so decided on 3 visualizations

boroughs = airbnbnyc['neighbourhood_group'].unique()
boroughs
```

```
avgprice_borough = {}

for b in boroughs:
  avgprice_borough[b] = [airbnbnyc[airbnbnyc['neighbourhood_group'] == b]['price'].mean()]
print(avgprice_borough)
```

```
avg_price_b = pd.DataFrame(avgprice_borough).T.reset_index()

avg_price_b.columns=['Borough', 'Average Price']

avg_price_b.head()
```

```
# looking at the most expensive neighborhood in order to see if the really
# expesive places to stay get reviews
# removed 3 outliers in order to make the chart more readable 2 that were priced over 3500k per night
# one that had 450 reviews, much more than any other place

tribeca = manhattan[manhattan['neighbourhood'] == "Tribeca"]
outliers = [2163, 30268, 48043]

tribeca = tribeca.drop(outliers)
```

```
  import altair as alt

ichart = alt.Chart(tribeca, title='Number of Reviews compared to Price per Night: Tribeca').mark_point().encode(
    x=alt.X('price', title='Price per Night ($)'),
    y=alt.Y('number_of_reviews', title='Number of Reviews'),
    color='room_type',
    tooltip=['price','number_of_reviews','room_type','host_name', 'name', 'latitude', 'longitude']
).interactive()

ichart
```

```
# exporting chart
ichart.save('tribecreviewsprice.html')
```
